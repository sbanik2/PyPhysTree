```python
import numpy as np
import random
import logging
from lgtree.lgtree  import  run_optimisation_mcts
import matplotlib.pyplot as plt
```

```python
# plotting
```

```python
# -------------------- 2. Plot 3D Surface --------------------
def plot_3d_function(func, lb, ub, resolution=100):
    x = np.linspace(lb[0], ub[0], resolution)
    y = np.linspace(lb[1], ub[1], resolution)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(resolution):
        for j in range(resolution):
            _, Z[i, j] = func(np.array([X[i, j], Y[i, j]]))

    fig = plt.figure(figsize=(4, 4))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')
    ax.set_xlabel("x1")
    ax.set_ylabel("x2")
    ax.set_zlabel("f(x)")
    plt.tight_layout()
    plt.show()
```

```python
# Sphere
```

```python
def sphere(x):
    """
    F1  Sphere Function
    f(x) = sum(x_i^2)
    - Bounds: [-100, 100]
    - Dimension: any (here: 2 for plotting)
    - Global Minimum: f(x) = 0 at x = 0
    """
    return x, np.sum(x**2)


# -------------------- 3. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 4. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 5. Define Objective and Bounds --------------------
dim = 2
objfunc =  sphere
lb = np.array([-100] * dim)
ub = np.array([100] * dim)

# -------------------- 6. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 7. Run 3D Visualization --------------------
plot_3d_function(sphere, lb, ub)

# -------------------- 8. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```

```python
# -------------------- 1. Define Ackley Function --------------------
def ackley(x):
    """
    F10  Ackley Function
    f(x) = -20 * exp(-0.2 * sqrt(1/n * sum(x_i^2))) 
           - exp(1/n * sum(cos(2x_i))) + 20 + e
    - Bounds: [-32, 32]
    - Global Minimum: f(x) = 0 at x = 0
    """
    n = len(x)
    sum_sq = np.sum(x**2)
    sum_cos = np.sum(np.cos(2 * np.pi * x))
    return x, -20 * np.exp(-0.2 * np.sqrt(sum_sq / n)) - np.exp(sum_cos / n) + 20 + np.e

# -------------------- 2. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 3. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 4. Define Objective and Bounds --------------------
dim = 2
objfunc = ackley
lb = np.array([-32] * dim)
ub = np.array([32] * dim)

# -------------------- 5. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 6. Run 3D Visualization --------------------
plot_3d_function(ackley, lb, ub)

# -------------------- 7. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```

```python
# -------------------- 1. Define Penalized Function 1 --------------------
def penalized1(x):
    """
    F12  Penalized Function 1
    f(x) = /n * {10*sin(*y1) + sum((y_i - 1)^2 * [1 + 10sin(*y_{i+1})]) + (y_n - 1)^2}
           + sum(u(x_i, 10, 100, 4))
    where y_i = 1 + (x_i + 1)/4
          u(x_i, a, k, m) is a penalty function
    - Bounds: [-50, 50]
    - Global Minimum: f(x) = 0
    """
    def u(xi, a=10, k=100, m=4):
        if xi > a:
            return k * (xi - a)**m
        elif xi < -a:
            return k * (-xi - a)**m
        else:
            return 0

    y = 1 + (x + 1) / 4
    term1 = 10 * np.sin(np.pi * y[0])**2
    term2 = np.sum((y[:-1] - 1)**2 * (1 + 10 * np.sin(np.pi * y[1:])**2))
    term3 = (y[-1] - 1)**2
    penalty = np.sum([u(xi) for xi in x])
    return x, (np.pi / len(x)) * (term1 + term2 + term3) + penalty

# -------------------- 2. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 3. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 4. Define Objective and Bounds --------------------
dim = 2
objfunc = penalized1
lb = np.array([-50] * dim)
ub = np.array([50] * dim)

# -------------------- 5. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 6. Run 3D Visualization --------------------
plot_3d_function(penalized1, lb, ub)

# -------------------- 7. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```

```python
# -------------------- 1. Define Shekels Foxholes (dejongN5) Function (F14) --------------------
def shekel_foxholes(x):
    """
    F14  Shekels Foxholes (De Jongs Function No. 5)
    f(x) = [0.002 + sum_{i=1}^{25} 1 / (i + (x1 - a1_i)^6 + (x2 - a2_i)^6)]^(-1)
    - Bounds: [-65, 65]
    - Dimension: 2
    - Global Minimum: f(x)  1 at x = [-32, -32]
    """
    A = [-32., -16., 0., 16., 32.]
    a1 = np.array(A * 5)
    a2 = np.repeat(A, 5)
    val = 0
    for i in range(25):
        val += 1 / (i + 1 + (x[0] - a1[i])**6 + (x[1] - a2[i])**6)
    return x, (0.002 + val)**(-1)

# -------------------- 2. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 3. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 4. Define Objective and Bounds --------------------
dim = 2
objfunc = shekel_foxholes
lb = np.array([-65.0] * dim)
ub = np.array([65.0] * dim)

# -------------------- 5. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 6. Run 3D Visualization --------------------
plot_3d_function(shekel_foxholes, lb, ub)

# -------------------- 7. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```

```python
# -------------------- 1. Define Kowalik Function --------------------
def kowalik(x):
    """
    F15  Kowalik Function
    f(x) = sum_{i=1}^{11} [a_i - (x1(bi^2 + bi*x2)) / (bi^2 + bi*x3 + x4)]^2
    - Bounds: [-5, 5]
    - Dimension: 4
    - Global Minimum:  0.000307 at x  [0.1928, 0.1908, 0.1231, 0.1358]
    """
    b = np.array([4.0, 2.0, 1.0, 1/2.0, 1/4.0,
                  1/6.0, 1/8.0, 1/10.0, 1/12.0, 1/14.0, 1/16.0])
    a = np.array([0.1957, 0.1947, 0.1735, 0.1600, 0.0844,
                  0.0627, 0.0456, 0.0342, 0.0323, 0.0235, 0.0246])

    val = 0.0
    for i in range(11):
        bi = b[i]
        num = x[0] * (bi**2 + bi * x[1])
        denom = bi**2 + bi * x[2] + x[3]
        val += (a[i] - num / denom) ** 2
    return x, val

# -------------------- 2. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 3. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 4. Define Objective and Bounds --------------------
dim = 4
objfunc = kowalik
lb = np.array([-5.0] * dim)
ub = np.array([5.0] * dim)

# -------------------- 5. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 6. Run 3D Visualization --------------------
# Optional for dim > 2, you can skip or adapt visualization

# -------------------- 7. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```

```python
# -------------------- 1. Define Hartmann 3D Function and Run MCTS --------------------
def hartmann_3d(x):
    """
    F19  Hartmann 3D Function
    f(x) = -sum_{i=1}^4 _i * exp(-sum_{j=1}^3 A_ij * (x_j - P_ij)^2)
    - Bounds: [0, 1]
    - Dimension: 3
    - Global Minimum:  -3.86278 at specific x
    """
    alpha = np.array([1.0, 1.2, 3.0, 3.2])
    A = np.array([
        [3.0, 10, 30],
        [0.1, 10, 35],
        [3.0, 10, 30],
        [0.1, 10, 35]
    ])
    P = 1e-4 * np.array([
        [3689, 1170, 2673],
        [4699, 4387, 7470],
        [1091, 8732, 5547],
        [381, 5743, 8828]
    ])
    
    outer = 0.0
    for i in range(4):
        inner = np.sum(A[i] * (x - P[i])**2)
        outer += alpha[i] * np.exp(-inner)
    return x, -outer

# -------------------- 2. Set Global Seed --------------------
seed_value = random.randint(1, 100000)

# -------------------- 3. Setup Logging --------------------
logging.basicConfig(filename='mcts_log.txt', level=logging.INFO, filemode='w')
logger = logging.getLogger('MCTSLogger')

# -------------------- 4. Define Objective and Bounds --------------------
dim = 3
objfunc = hartmann_3d
lb = np.array([0.0] * dim)
ub = np.array([1.0] * dim)

# -------------------- 5. Global MCTS Batch Setup --------------------
ntrees = 20
a_min, a_max = 0.008, 0.08
explore_constant_min = 1e-1
top_k = 5  # number of best candidates for local evaluation

# -------------------- 6. Run 3D Visualization --------------------
#plot_3d_function(hartmann_3d, lb, ub)

# -------------------- 7. Optimization Run --------------------
minscore, best_parameter = run_optimisation_mcts(
    objfunc=objfunc,
    logger=logger,
    seed_value=seed_value,
    lb=lb,
    ub=ub,
    ntrees=ntrees,
    top_k=top_k,
    a_min=a_min,
    a_max=a_max,
    explore_constant_min=explore_constant_min,
)
```
